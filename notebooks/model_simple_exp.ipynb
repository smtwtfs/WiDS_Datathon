```python
import pandas as pd
import xgboost as xgb
import numpy as np
```

```python
df_res = pd.read_excel("/content/drive/MyDrive/WiDS_datathon_2025/TRAIN/TRAINING_SOLUTIONS.xlsx")
df_q = pd.read_excel("/content/drive/MyDrive/WiDS_datathon_2025/TRAIN/TRAIN_QUANTITATIVE_METADATA.xlsx")

df_all = df_res.merge(df_q, on = "participant_id")
y_str = pd.Series([str(s) + str(a) for s, a in zip(df_all['ADHD_Outcome'], df_all['Sex_F'])])
y_bin = pd.Series([int(str(s) + str(a),2) for s, a in zip(df_all['ADHD_Outcome'], df_all['Sex_F'])])
```

```python
X = df_all.drop(columns= ['participant_id', 'ADHD_Outcome', 'Sex_F'])
y = y_bin.astype("category")

#dtrain = xgb.DMatrix(X, y, enable_categorical = True)
```

```python
# Supported tree methods are `approx` and `hist`.
clf = xgb.XGBClassifier(tree_method="hist", enable_categorical=True, device="cuda")
# X is the dataframe we created in previous snippet
clf.fit(X, y)
```
