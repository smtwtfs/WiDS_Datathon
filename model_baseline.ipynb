```python
import pandas as pd
import xgboost as xgb
import numpy as np
```

```python
df_res = pd.read_excel("/content/drive/MyDrive/WiDS_datathon_2025/TRAIN/TRAINING_SOLUTIONS.xlsx")
df_q = pd.read_excel("/content/drive/MyDrive/WiDS_datathon_2025/TRAIN/TRAIN_QUANTITATIVE_METADATA.xlsx")
```

```python
df_all = df_res.merge(df_q, on = "participant_id")
```

```python
df_all
```

```python
X = df_all.drop(columns= ['participant_id', 'ADHD_Outcome', 'Sex_F'])
y_s = df_all['Sex_F']
y_a = df_all['ADHD_Outcome']
```

```python
dtrain_s = xgb.DMatrix(X, y_s)
dtrain_a = xgb.DMatrix(X, y_a)
```

```python
model_s = xgb.train(params = {}, dtrain = dtrain_s)
model_a = xgb.train(params = {}, dtrain = dtrain_a)
```

```python
# pred
```

```python
test_df = pd.read_excel('/content/drive/MyDrive/WiDS_datathon_2025/TEST/TEST_QUANTITATIVE_METADATA.xlsx')
test_df.head()
```

```python
dtest = xgb.DMatrix(test_df.drop(columns = ['participant_id']))
```

```python
(df_all['ADHD_Outcome'])
```

```python
pred_s = model_s.predict(dtest)
pred_a = model_a.predict(dtest)
```

```python
our_res = pd.DataFrame({'participant_id': test_df.participant_id, 'ADHD_Outcome': (pred_a > 0.5).astype(int), 'Sex_F': (pred_s > 0.5).astype(int)})
```
